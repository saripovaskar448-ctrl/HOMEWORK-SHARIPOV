```python
#домашнняя работа усложненная
day_weather_10_years = [17, 8, 4, 24, 9, 12, 13, 15, 19, 14,]
summa_pogodi = sum(day_weather_10_years)
goda = len(day_weather_10_years)
average = summa_pogodi / goda
print("cредняя темпа составляет:", average)

pogoda_25_god = 12
sravnenie = average > pogoda_25_god

print("средняя погода больше чем в 25 году :", sravnenie)

if average > pogoda_25_god:
    raznica = pogoda_25_god - average
    print("температура выше градусов на :", raznica)
    print("температура больше прогноза")
elif average < pogoda_25_god:
    raznica12 = pogoda_25_god - average
    print("Температура ниже градусов на :", raznica12)
    print("температура меньше прогноза")
```
