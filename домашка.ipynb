```python
word = input("Введите слово: ")

bykv = len(word)

if bykv % 2 == 0:
    #четное
    seredina = bykv // 2
    print(word[seredina-1] + word[seredina])
else:
    #нечетное
    seredina = bykv // 2
    print(word[seredina])
```

```python
chislo = 0

while True:
    num = int(input("Введите число : "))
    if num == 0:
        break
    chislo += num
print("Сумма : ", chislo)
```

```python
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']

if len(boys) != len(girls):
    print ("кто то может остаться без пары!")
else:
    sort_boy = sorted(boys)
    sort_girl = sorted(girls)

print('Идеальные пары :')
print(f"{sort_boy} и {sort_girl}")
```

```python
countries_temperature = [
    ['Таиланд', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
    ['Германия', [57.2, 55.4, 59, 59, 53.6]],
    ['Россия', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
    ['Польша', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]

for country in countries_temperature:
    name = country[0]
    temperatyra = country[1]
    
    #сумма темп
    total = 0
    for temp in temperatyra:
        total += temp
    
    #ср.темп
    srednya_f = total / len(temperatyra)
    
    # цельси
    celsii = (srednya_f - 32) * 5 / 9

    print(f"{name}: {celsii} C")
```
